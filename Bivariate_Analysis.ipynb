```python
df3['Mother.edu'].value_counts()
```

```python
df3['Siblings'].value_counts()
```

```python
def siblings(x):
    if '2' in x:
        return 2
    if '1' in x:
        return 1
    if '3' in x:
        return 3
    if '4' in x:
        return 4
    if x=='Single child':
        return 0
```

```python
df3['Siblings_count'] = df3['Siblings'].apply(lambda x: siblings(x))
df3[['Siblings','Siblings_count']].head(20)
```

```python
df3.groupby('Mother.edu')['Siblings_count'].sum()
```

```python
df3.groupby(['Mother.edu','Siblings'])['Siblings_count'].sum()
```

```python
df3.groupby(['Mother.edu','Siblings'])['Siblings'].count()
```
